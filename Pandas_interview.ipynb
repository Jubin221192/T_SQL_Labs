# How to combine skiprows and nrows to read data in small chunks

```python
li = []

no_of_rows = 10

chunks =5

df = pd.read_csv("Boston_housing.csv",nrows=2)
```

```python
colnames = df.columns

for i in range(0, no_of_rows * chunks, no_of_rows):
    
    df = pd.read_csv("Boston_housing.csv", header = 0, skiprows = i,nrows = no_of_rows, names = colnames)
    
    li.append(df)
```

```python
len(li)
```

```python
import csv
from collections import namedtuple

Msg = namedtuple('Msg', ('Bedroom', 'Sqfoot', 'Loc', 'Price'))

messages = []
with open('CSV_EX_2.csv') as csvfile:
    msgreader = csv.reader(csvfile)
    for row in msgreader:
        messages.append(Msg(*row))
```

```python
messages
```

# Working On Ordered dictionary

```python
from collections import OrderedDict
b=OrderedDict()
b['egg'] = 2
b['fish'] = 3
b['milk'] = 1
b['paneer'] =4
b
```

```python
for key, value in b.items():
    print('key:{0}   Value:{1}'.format(key,value))
```

```python
# updating the ordered list
b.update([('egg', 77),('paneer', 7)])
```

```python
# adding new keys to the data
b['chicken'] = 12
```

```python
# fetching data using get function
b.get('chicken', 'Not Found')
```

```python
# deleting keys from the ordered list

x = b.pop('chicken')
```
